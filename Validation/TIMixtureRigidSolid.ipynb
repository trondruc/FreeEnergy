# Thermodynamic integration for binary mixtures

The standard relation is

\begin{align}
p = -\Big(\frac{\partial A}{\partial V}\Big)_{N,T},\\
\end{align}

The partition function for a binary mixture is given by

\begin{align}
Z = \frac{1}{\Lambda^{3N}}\frac{1}{N_{A}!N_{B}!}\int \exp\big[-U(\textbf{r}^{N})/k_{\rm B} T\big]d\textbf{r}^{N} = Z_{\rm id-mix}Z_{\rm ex},
\end{align}

with $\Lambda = h/\sqrt{2\pi m k_{\rm B}T}$ and

\begin{align}
& Z_{id-mix} = \frac{1}{\Lambda^{3N}}\frac{1}{N_{\rm A}!N_{\rm B}!}V^{N},\\
& Z_{\rm ex} = V^{-N}\int \exp\big[-U(\textbf{r}^{N})/k_{\rm B} T\big]d\textbf{r}^{N}.
\end{align}

Hence

\begin{align}
A(\rho,T) & = A_{\rm id-mix}(\rho,T) + A_{\rm ex}(\rho,T),\\
\beta A_{\rm id-mix}(\rho,T) & = N\big(3 \ln\Lambda + \ln\rho - 1\big) - \ln\Big(\frac{N!}{N_{\rm A}!N_{\rm B}!}\Big)
\end{align}

Inserting we get

\begin{align}
-p = \Big(\frac{\partial A_{\rm id-mix}}{\partial V}\Big)_{N,T} + \Big(\frac{\partial A_{\rm ex}}{\partial V}\Big)_{N,T}.
\end{align}

And therefore

\begin{align}
W = -V\Big(\frac{\partial A_{\rm ex}}{\partial V}\Big)_{N,T}.
\end{align}

Inserting $-d\ln V = -(N/V)d(V/N) = - \rho d(\rho^{-1}) = \rho (\rho^{-2})d\rho = \rho^{-1}d\rho = d\ln \rho$ we get

\begin{align}
W = \Big(\frac{\partial A_{\rm ex}}{\partial \ln \rho}\Big)_{N,T}.
\end{align}

We have $A = E - TS$ and thus using $S = - (\partial A/\partial T)_{N,V}$ we get the Gibbs-Helmholtz equation

\begin{align}
E = \Big(\frac{\partial (\beta A)}{\partial \beta} \Big)_{N,V},
\end{align}

and therefore $\big(A_{\rm ex} = U_{\rm ex} - TS_{\rm ex}$, $S_{\rm ex} = - (\partial A_{\rm ex}/\partial T)_{N,V}\big)$

\begin{align}
U_{ex} = \Big(\frac{\partial (\beta A_{ex})}{\partial \beta} \Big)_{N,V}.
\end{align}

Other useful relations

\begin{align}
pV = \Big(\frac{\partial A}{\partial\ln \rho}\Big)_{N,T},\\
\frac{\partial}{\partial \rho}\Big(\frac{\beta A}{N}\Big)_{N,T} = \frac{Z}{\rho}.
\end{align}

with $Z = P/(\rho k_{\rm B}T)$. For single-component systems:

\begin{align}
\mu_{\rm ex} = \frac{G_{\rm ex}}{n} = \frac{U_{\rm ex} + W - TS_{\rm ex}}{n} = \frac{A_{\rm ex} + W}{n}.
\end{align}

# Thermodynamic integration for rigid molecules

Again

\begin{align}
p = -\Big(\frac{\partial A}{\partial V}\Big)_{N,T}.
\end{align}

The partition function for a system of rigid molecules is given by

\begin{align}
Z = \frac{(\Lambda_{x} \Lambda_{y} \Lambda_{z})^{N}}{\lambda^{3N}}\frac{1}{N!}\int \exp\big[-U(\textbf{q}^{N})/k_{\rm B} T\big]d\textbf{q}^{N} = Z_{\rm id}Z_{\rm ex},
\end{align}

with $\Lambda_{x} = (2\pi I_{x} k_{\rm B}T)^{1/2}/h$, and $\lambda = h/\sqrt{2\pi m k_{\rm B}T}$. Hence

\begin{align}
A(\rho,T) = A_{\rm id}(\rho,T) + A_{\rm ex}(\rho,T).
\end{align}

Inserting we get

\begin{align}
-p = \Big(\frac{\partial A_{\rm id}}{\partial V}\Big)_{N,T} + \Big(\frac{\partial A_{\rm ex}}{\partial V}\Big)_{N,T}.
\end{align}

The rigid bonds in simulations give a contribution to $W$ and therefore

\begin{align}
W - W_{\rm id} = -V\Big(\frac{\partial A_{\rm ex}}{\partial V}\Big)_{N,T}.
\end{align}

Inserting $-d\ln V = d\ln \rho$ we get

\begin{align}
W - W_{\rm id} = \Big(\frac{\partial A_{\rm ex}}{\partial \ln \rho}\Big)_{N,T}.
\end{align}

This equation can then be integrated from low density (ideal molecular gas). We can find $W_{\rm id}$ from

\begin{align}
W_{\rm id} = T (n -  g /3), 
\end{align}

since $p_{\rm id} = \rho k_{\rm B} T = (2/3K + W_{\rm id})/V$.

# Thermodynamic integration for solids

The potential for an Einstein solid is (no static term)

\begin{align}
V = \sum_{i=1}^{N} k\big(\textbf{r}_{i} - \textbf{r}_{i,0}\big)^{2}.
\end{align}

The partition function for a fixed crystal lattice is ($\textbf{s}_{i} = \textbf{r}_{i} - \textbf{r}_{i,0}$ and no $N!$ factor since fixed)

\begin{align}
Q & = \frac{1}{\Lambda^{3N}}\int_{1} \exp\big[-U(\textbf{r}^{N})/k_{\rm B} T\big]d\textbf{r}^{N},\\ 
& = \frac{1}{\Lambda^{3N}}\int \exp\big[-\sum k\beta \textbf{s}_{i}^{2}\big]d\textbf{s}^{N},\\
& = \frac{1}{\Lambda^{3N}}\Big(\frac{\pi}{k\beta}\Big)^{3N/2}.
\end{align}

as the Gaussian integral can be performed independently for each particle. For a system with fixed center-of-mass and conserved momentum $N$ is replaced 
with $N-1$. The absolute free energy becomes

\begin{align}
A\beta = - \ln\Big(\frac{\pi}{k\beta}\Big)^{3N/2} + 3N \ln \Lambda.
\end{align}
